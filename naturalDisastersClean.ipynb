```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('Resources/data/1990_2022_natural_disasters.csv')
df.head()
```

```python
list(df)
```

```python
df = df.drop(columns=["Glide", "Disaster Subsubtype", "ISO", "Origin", "Associated Dis", "Associated Dis2","OFDA Response", "Appeal", "Declaration", "Local Time", "River Basin", "No Injured", "No Affected", "No Homeless", "Adm Level", "Admin1 Code", "Admin2 Code"])
df.head()
```

```python
df.columns=df.columns.str.strip().str.lower()
df.columns
```

```python
df2= df.rename(str.strip, axis ='columns')
df2
```

```python
df2.columns
```

```python
df3= df2.drop(df2[df2['disaster group'] == 'Technological'].index)

print(df3)
```

```python
df4= df3.drop(df3[df3['disaster group'] == 'Complex Disasters'].index)                                 
```

```python
df4.columns=df4.columns.str.replace(' ','_')
df4=df4.fillna(0)
city_cleanup=df4['location'].str.split(',', expand=True)
city_cleanup=city_cleanup[0].fillna(0)
df4['location_estimate']=city_cleanup
df4=df4.drop(columns=['cpi','geo_locations','latitude', 'longitude'])
```

```python
df4['start_month']=df4['start_month'].replace(0,1).map("{:.1}".format)
df4['start_day']=df4['start_day'].replace(0,1).map("{:.1}".format)
df4['end_month']=df4['end_month'].replace(0,1).map("{:.1}".format)
df4['end_day']=df4['end_day'].replace(0,1).map("{:.1}".format)
df4['start_date']=df4['start_year'].astype('str')+'-'+df4['start_month']+'-'+df4['start_day']
df4['end_date']=df4['end_year'].astype('str')+'-'+df4['end_month']+'-'+df4['end_day']
df4['start_date']=pd.to_datetime(df4['start_date'])
df4['end_date']=pd.to_datetime(df4['end_date'])
```

```python
for index,row in df4.iterrows():
    country=row["country"]
    if row['location_estimate']==0:
        df4.loc[index,'location_estimate']=country
```

```python
df4
```

```python
import requests
from config import gmap_key
```

```python
df4['lat']=''
df4['lng']=''
for index,row in df4.iterrows():
    try:
        target=row['location_estimate']
        url = f"https://maps.googleapis.com/maps/api/geocode/json?address={target}&key={gmap_key}"
        response = requests.get(url).json()
        lat = response['results'][0]['geometry']['location']['lat']
        lng = response['results'][0]['geometry']['location']['lng']
        df4.loc[index,'lat']=lat
        df4.loc[index,'lng']=lng
    except:
        target=row['country']
        url = f"https://maps.googleapis.com/maps/api/geocode/json?address={target}&key={gmap_key}"
        response = requests.get(url).json()
        lat = response['results'][0]['geometry']['location']['lat']
        lng = response['results'][0]['geometry']['location']['lng']
        df4.loc[index,'lat']=lat
        df4.loc[index,'lng']=lng
```

```python
df4.to_csv('Resources/data/natural_disasters.csv')
```
