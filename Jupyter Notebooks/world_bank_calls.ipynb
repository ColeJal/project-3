```python
import pandas as pd
import requests
```

```python
country_code_list = pd.read_csv('Resources/data/WDICountry.csv')
code_list=country_code_list[["Country Code","Table Name"]]
code_list=code_list.rename(columns={'Country Code':'country_code','Table Name':'Country'})
code_list
```

```python
url = "http://api.worldbank.org/v2/country/all/indicator"
api_format = "json"
gdp_collection = []
world_response = requests.get(f"{url}/NY.GDP.PCAP.CD?date=2018&format={api_format}").json()
total_pages = world_response[0]['pages']+1
for page in range(total_pages):
    response=requests.get(f"{url}/NY.GDP.PCAP.CD?date=2018&format={api_format}&page={page}").json()
    if page != 0:
        gdp_collection.append(response[1])
    else:
        pass
gdp_list=[]
for i in range(len(gdp_collection)):
    for j in range(len(gdp_collection[i])):
        country_code = gdp_collection[i][j]['countryiso3code']
        gdp = gdp_collection[i][j]['value']
        gdp_dict = {'country_code':country_code,
        'gdp_per_capita':gdp}
        gdp_list.append(gdp_dict)
gdp_df=pd.DataFrame(gdp_list)
gdp_2018 = pd.merge(gdp_df,code_list,on='country_code',how='left')
gdp_2018.to_csv('Resources/data/world_bank/gdp_2018.csv')
```

```python
url = "http://api.worldbank.org/v2/country/all/indicator"
api_format = "json"
co2_collection=[]
co2_response = requests.get(f"{url}/EN.ATM.CO2E.KT?date=2018&format={api_format}").json()
total_pages = co2_response[0]['pages']+1
for page in range(total_pages):
    response=requests.get(f"{url}/EN.ATM.CO2E.KT?date=2018&format={api_format}&page={page}").json()
    if page != 0:
        co2_collection.append(response[1])
    else:
        pass
co2_list=[]
for i in range(len(co2_collection)):
    for j in range(len(co2_collection[i])):
        country_code = co2_collection[i][j]['countryiso3code']
        co2 = co2_collection[i][j]['value']
        co2_dict = {'country_code':country_code,
        'total_co2':co2}
        co2_list.append(co2_dict)
co2_df=pd.DataFrame(co2_list)
co2_2018 = pd.merge(co2_df,code_list,on='country_code',how='left')
co2_2018.to_csv('Resources/data/world_bank/total_co2_2018.csv')
country_stats2018 = pd.merge(gdp_df,co2_df,on='country_code',how='left')
```

```python
url = "http://api.worldbank.org/v2/country/all/indicator/EN.ATM.CO2E.PC"
api_format = "json"
co2_per_cap_collection=[]
co2_per_cap_response = requests.get(f"{url}?date=2018&format={api_format}").json()
total_pages = co2_per_cap_response[0]['pages']+1
for page in range(total_pages):
    response=requests.get(f"{url}?date=2018&format={api_format}&page={page}").json()
    if page != 0:
        co2_per_cap_collection.append(response[1])
    else:
        pass
co2_per_cap_list=[]
for i in range(len(co2_per_cap_collection)):
    for j in range(len(co2_per_cap_collection[i])):
        country_code = co2_per_cap_collection[i][j]['countryiso3code']
        co2_per_cap = co2_per_cap_collection[i][j]['value']
        co2_dict = {'country_code':country_code,
        'co2_per_cap':co2_per_cap}
        co2_per_cap_list.append(co2_dict)
co2_per_cap_df=pd.DataFrame(co2_per_cap_list)
co2_per_cap_2018 = pd.merge(co2_per_cap_df,code_list,on='country_code',how='left')
co2_per_cap_2018.to_csv('Resources/data/world_bank/co2_per_cap_2018.csv')
country_stats2018 = pd.merge(country_stats2018,co2_per_cap_df,on='country_code',how='left')
```

```python
url = "http://api.worldbank.org/v2/country/all/indicator/SP.POP.TOTL"
api_format = "json"
population_collection=[]
population_response = requests.get(f"{url}?date=2018&format={api_format}").json()
total_pages = population_response[0]['pages']+1
for page in range(total_pages):
    response=requests.get(f"{url}?date=2018&format={api_format}&page={page}").json()
    if page != 0:
        population_collection.append(response[1])
    else:
        pass
population_list=[]
for i in range(len(population_collection)):
    for j in range(len(population_collection[i])):
        country_code = population_collection[i][j]['countryiso3code']
        pop = population_collection[i][j]['value']
        pop_dict = {'country_code':country_code,
        'population':pop}
        population_list.append(pop_dict)
population_df=pd.DataFrame(population_list)
population_2018 = pd.merge(population_df,code_list,on='country_code',how='left')
population_2018.to_csv('Resources/data/world_bank/population_2018.csv')
country_stats2018 = pd.merge(country_stats2018,population_df,on='country_code',how='left')
```

```python
country_stats2018 = pd.merge(country_stats2018,code_list,on='country_code',how='left')
country_stats2018=country_stats2018.loc[country_stats2018['country_code']!="",:]
country_stats2018.to_csv('Resources/cleaned/country_stats2018.csv')
```
