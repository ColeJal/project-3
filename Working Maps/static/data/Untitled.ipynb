```python
import pandas as pd
```

```python
disasters=pd.read_csv('natural_disasters.csv')
disasters
```

```python
summary=disasters[['country','year','disaster_type','disaster_subtype','total_deaths', 'total_affected','start_date'
       ]]
```

```python
disasters.columns
```

```python
totals=summary.groupby(['year','disaster_type']).sum()
totals['total_occurences']=summary.groupby(['year','disaster_type']).count()['start_date']
totals.to_csv('yearlybreakdowns.csv')
```
